```python
import pandas as pd
import os
import numpy as np

dataset = pd.read_csv('labels.csv')
```

```python
print(dataset.shape)
```

```python
image_id=(dataset['id']+".jpg")
breeds=dataset['breed']
```

```python
print(breeds)
```

```python
def search(element,breed_lists):
    for breed_list in breed_lists:
        if (element == breed_list):
            return False
    return True
```

Make folders with dog breed names

```python
import cv2
breed_lists=[]
path = './breeds'
for x in range (10222):
    while search(breeds[x],breed_lists):
        breed_lists.append(breeds[x])
        image= cv2.imread(os.path.join('train',image_id[x]))
        os.mkdir(os.path.join(path,breeds[x]))
        try:
            cv2.imwrite(os.path.join(path,breeds[x],image_id[x]),image)
        except Exception as e:
            print(e)

     
```

```python
from keras.utils.np_utils import to_categorical
path = './breeds'

categories=os.listdir(path)
print(categories)

labels=np.arange(len(categories))
print(labels)

labels_new = []
labels_new=to_categorical(labels)
print (labels_new)

label_dict=dict(zip(categories,labels_new))
print(label_dict)
```

Preprocess

```python
import cv2

img_size=100

data=[]
target=[]

for category in categories:
    
    imgs_path=os.path.join(path,category)
    #print(imgs_path)
    img_names=os.listdir(imgs_path)
    #print(img_names)
    
    for img_name in img_names:
        img_path=os.path.join(imgs_path,img_name)
        #print(img_path)
        try:
            
            img=cv2.imread(img_path)
            img=cv2.resize(img,(img_size,img_size))

            data.append(img_path)
            target.append(label_dict[category])
        
        except Exception as e:
            
            print(e)
            
    #     cv2.imshow('LIVE',img)
    #     k=cv2.waitKey(1000)
    #     if(k==27):
    #         break
    # cv2.destroyAllWindows()    
```
